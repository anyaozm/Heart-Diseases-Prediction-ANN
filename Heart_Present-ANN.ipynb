```python
#Import dependencies
import numpy as np
import pandas as pd
import keras
import tensorflow as tf
import matplotlib.pyplot as plt
from matplotlib import style
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn import preprocessing, svm, metrics  
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation, Flatten
from keras.layers import Embedding
import math
from keras import metrics
```

```python
#Load Training Features Data
train_v = '/Users/anyaozmen/Downloads/heart_warm/train_values.csv'
```

```python
#Load Training Label Data
train_l = '/Users/anyaozmen/Downloads/heart_warm/train_labels.csv'
```

```python
#Turn data into a pandas data frame and Display
df_tval = pd.read_csv(train_v, index_col=0)
#df_tval
```

```python
#Turn data into a pandas data frame and Display
df_tlab = pd.read_csv(train_l, index_col=0)
#df_tlab
```

```python
#Replace the strings in thal column with corresponding numbers
df_tval = df_tval.replace({'normal': 2, 'reversible_defect': 1, 'fixed_defect': 0})
```

```python
#Turn features into numpy arrays  
X = np.array(df_tval)
X = preprocessing.scale(X)
```

```python
y = np.array(df_tlab)
```

```python
#Check shape
X.shape
```

```python
#Split the model 
X_train, X_test, y_train, y_test =  train_test_split(X, y, test_size=0.17, random_state=13)
```

```python
#Build the Model 
model = Sequential()
model.add(Dense(4, input_shape=(13,), activation='relu'))
model.add(Dense(6, input_shape=(13,), activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(2, input_shape=(13,), activation='relu'))
model.add(Dropout(0.4))
model.add(Dense(10, input_shape=(13,), activation='relu'))
model.add(Dropout(0.4))
model.add(Dense(1, activation='sigmoid'))
```

```python
#Compile
model.compile(optimizer='Adam', loss= 'binary_crossentropy', metrics=['accuracy'])
```

```python
#Fit and train
model.fit(X_train, y_train, batch_size = 32, epochs=600)
```

```python
#Let's Evaluate our model
score = model.evaluate(X_test, y_test)
print(score)
#len(y_test)
```
