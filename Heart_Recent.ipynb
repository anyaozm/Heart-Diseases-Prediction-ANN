```python
import numpy as np
import pandas as pd
import keras
import tensorflow as tf
import matplotlib.pyplot as plt
from matplotlib import style
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn import preprocessing, svm, metrics  
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation, Flatten
from keras.layers import Embedding
import math
from keras import metrics
```

```python
train_v = '/Users/anyaozmen/Downloads/heart_warm/train_values.csv'
```

```python
train_l = '/Users/anyaozmen/Downloads/heart_warm/train_labels.csv'
```

```python
df_tval = pd.read_csv(train_v, index_col=0)
#df_tval
```

```python
df_tlab = pd.read_csv(train_l, index_col=0)
#df_tlab
```

```python
#replace the strings in thal column with corresponding numbers
df_tval = df_tval.replace({'normal': 2, 'reversible_defect': 1, 'fixed_defect': 0})
```

```python
X = np.array(df_tval)
X = preprocessing.scale(X)
```

```python
y = np.array(df_tlab)
```

```python
X.shape
```

```python
X_train, X_test, y_train, y_test =  train_test_split(X, y, test_size=0.20, random_state=13)
```

```python
model = Sequential()
model.add(Dense(4, input_shape=(13,), activation='relu'))
model.add(Dense(200, input_shape=(13,), activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(2, input_shape=(13,), activation='relu'))
model.add(Dropout(0.4))
model.add(Dense(10, input_shape=(13,), activation='relu'))
model.add(Dropout(0.4))
model.add(Dense(8, input_shape=(13,), activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(optimizer='Adam', loss= 'binary_crossentropy', metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, batch_size = 32, epochs=300)
```

```python
#Let's Evaluate our model
score = model.evaluate(X_test, y_test)
print(score)
```

```python
y_pred = model.predict([X_test])
#y_pred = y_pred.round(2)
```

```python
sklearn.metrics.log_loss(y_test, y_pred, normalize=True, sample_weight=None, labels=None)
```
